Cleaning of the Dataset

```python
backup = car.copy()
```

```python
car = car[car['year'].astype(str).str.isnumeric()]
car['year'] = car['year'].astype(int)
```

```python
car['year'] = car['year'].astype(int)
```

```python
car = car[car['Price']!= "Ask For Price"]
```

```python
car['Price'] = pd.to_numeric(car['Price'].astype(str).str.replace(',', ''), errors='coerce')
```

```python
car.info()
```

```python
car.shape
```

```python
car['kms_driven'] = (
    car['kms_driven']
    .astype(str)                # ensure string type
    .str.split(' ').str.get(0)  # take only the numeric part
    .str.replace(',', '')       # remove commas
)

# finally convert to integer
car['kms_driven'] = car['kms_driven'].astype(int)
```

```python
car['kms_driven'] = (
    car['kms_driven']
    .astype(str)
    .str.replace(',', '')
    .str.extract('(\d+)')[0]    # extract digits only
)

car = car[car['kms_driven'].notnull()]  # drop non-numeric
car['kms_driven'] = car['kms_driven'].astype(int)
```

```python
car['kms_driven'] = car['kms_driven'].astype(int)
```

```python
car.info()
```

```python
car = car[~car['fuel_type'].isna()]
```

```python
car['name'] = car['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
car = car.reset_index(drop=True)
```

```python
car = car[car['Price']<6e6].reset_index(drop=True)
```

```python
x = car.drop(columns='Price')
y = car['Price']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(x[['name', 'company', 'fuel_type']])
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories = ohe.categories_),['name', 'company', 'fuel_type']), remainder='passthrough')
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans, lr)
```

```python
pipe.fit(X_train,Y_train)
```

```python
Y_pred = pipe.predict(X_test)
```

```python
r2_score(Y_test, Y_pred)
```

```python
scores = []
for i in range(1000):
    X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=i)
    lr = LinearRegression()
    pipe = make_pipeline(column_trans, lr)
    pipe.fit(X_train, Y_train)
    Y_pred = pipe.predict(X_test)
    scores.append(r2_score(Y_test, Y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=np.argmax(scores))
lr = LinearRegression()
pipe = make_pipeline(column_trans, lr)
pipe.fit(X_train, Y_train)
Y_pred = pipe.predict(X_test)
r2_score(Y_test, Y_pred)
```
